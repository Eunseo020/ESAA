### Step 1. Import the necessary libraries

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

### Step 2. Create the DataFrame that should look like the one below.
*참고*

```
raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
            'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
            'female': [0, 1, 1, 0, 1],
            'age': [42, 52, 36, 24, 73],
            'preTestScore': [4, 24, 31, 2, 3],
            'postTestScore': [25, 94, 57, 62, 70]}
```

````python
score_df = pd.DataFrame({'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
            'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
            'female': [0, 1, 1, 0, 1],
            'age': [42, 52, 36, 24, 73],
            'preTestScore': [4, 24, 31, 2, 3],
            'postTestScore': [25, 94, 57, 62, 70]})

score_df
````

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

````python
plt.figure(figsize=(5, 3))
sns.scatterplot(x='preTestScore', y='postTestScore', size='age', data=score_df)
plt.show()
````

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

````python
plt.figure(figsize=(5, 3))
sns.scatterplot(x='preTestScore', y='postTestScore', size=4.5*score_df['postTestScore'], hue=score_df['female'], data=score_df)
plt.show()
````

### BONUS: Create your own question and answer it.

preTestScore과 postTestScore의 성별 분포를 각각 violnplot으로 시각화하여 함께 나타내시오.

````python
fig, ax=plt.subplots(nrows=1, ncols=2, figsize=(7,5))
sns.violinplot(data=score_df, x='female', y='preTestScore', ax=ax[0])
sns.violinplot(data=score_df, x='female', y='postTestScore', ax=ax[1])
plt.show()
````
