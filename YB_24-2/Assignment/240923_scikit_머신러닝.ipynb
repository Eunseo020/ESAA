### **04. Model Selection 모듈 소개**
sklearn.model_selection        
(1) 학습/테스트 데이터 세트 분리
- 학습과 예측을 동일한 데이터 세트로 수행하면 예측 정확도가 100%. 따라서 예측은 다른 테스트 전용 데이터 세트로 해야함.
- train_test_split( 피처 데이터 세트, label 데이터 세트, 여러 파라미터...)
  - test_size: 테스트 데이터 크기
  - train_size: 학습 데이터 크기
  - shuffle: 분리 전 섞을지
  - random_state: 난수값 지정
  - 튜플 형태 반환
- .fit(X_train, y_train), .predict(X_test), accuracy_score(y_test, pred)

(2) 교차 검증
- 과적합(Overfitting), 성능 저하 문제-> 교차 검증으로 다양한 학습, 평가 수행
- 1차 교차검증 기반 평가, 2차 테스트 데이터로 평가 : 테스트 데이터셋 외 별도의 검증 데이터 사용(학습 데이터 셋에서 검증 데이터 셋을 추출)

*K 폴드 교차 검증*     
- K개의 데이터 폴드 세트를 만든 후 K번만큼 각 세트에 학습, 검증평가 반복 수행
- k=5일때: 5개의 폴드 데이터에서 각각 하나씩 검증세트일때의 학습, 검증 수행해 총 5개(K개)의 예측 평가를 구하고 평균냄 ==> 평가 결과
```
from sklearn.model_selection import KFold
KFold(n_splits=5) #k값 지정
```
- Stratified K 폴드: 불균형한 분포도(특정 레이블이 매우 많거나 적은)를 가진 레이블 데이터 집합을 위한 K폴드 방식.
  - 원본 데이터의 레이블 분포를 먼저 고려 후, 동일하게 학습/검증 데이터 셋 분배.
  - 회귀는 지원 안됨. 연속형 숫자값이므로 의미 없음
```
skf = StratifiedKFold(n_splits=3)
skf.split(iris_df, iris_df['label']) #레이블 데이터 세트가 인자로 필요
```

- cross_val_score(): 과정을 한번에 수행( 1.폴드 세트 설정, 2. for 루프로 학습/테스트 데이터 인덱스 반복 추출 3. 반복 학습, 예측 수행, 예측 성능 반환)
 - 내부적으로 StratifiedKFold 이용
> cross_val_score(estimator, X, y=None, cv=None, n_jobs=1, verbose=0, fit_params=None, pre_dispatch='2*n_jobs')
  - estimator: 분류/회귀 클래스
  - X= 피처 데이터 세트
  - y= 레이블 데이터 세트
  - scoring = 예측 성능 평가 기술('accuracy')
  - cv= 교차 검증 폴드 수
 - cv만큼 scoring파라미터로 평가 결괏값 배열로 반환

(3) GridSearchCV
- GridSearchCV API로 파라미터 입력, 파라미터 집합을 만들고 순차 적용하며 최적화 진행
- 교차검증 기반으로 순차적으로 파라미터의 최적 값을 찾음
- 수행시간 오래걸림
- cv=3이면 개별 파라미터 조합마다 3개의 폴딩세트를 3회에 걸쳐 학습/평가해 성능 측정
- estimator, param_grid, scoring, cv, refit(=True면 최적의 파라미터 저장)
- 딕셔너리 형태로 key에 하이퍼 파라미터 명칭, list로 파라미터 값 설정
- 일반적으로 GridSearchCV로 최적의 파라미터를 튜닝한 후 별도의 테스트 세트에서 평가함

````python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
import numpy as np

iris = load_iris()
features = iris.data
label = iris.target
dt_clf = DecisionTreeClassifier(random_state=156)

# 5개의 폴드 세트
kfold= KFold(n_splits=5)
cv_accuracy=[]
print('붓꽃 데이터 세트 크기:', features.shape[0])
````

````python
n_iter=0

# KFold 객체의 split() 호출로 학습, 검증용 테스트의 로우 인덱스를 array로 반환
for train_index, test_index in kfold.split(features):
  # kfold.split()으로 반환된 인덱스를 이용해 학습용, 검증용 테스트 데이터 추출
  X_train, X_test = features[train_index], features[test_index]
  y_train, y_test = label[train_index], label[test_index]
  # 학습 및 예측
  dt_clf.fit(X_train, y_train)
  pred = dt_clf.predict(X_test)
  n_iter += 1
  # 반복 시마다 정확도 측정
  accuracy = np.round(accuracy_score(y_test, pred), 4)
  train_size= X_train.shape[0]
  test_size= X_test.shape[0]
  print('\n#{0} 교차 검증 정확도 :{1}, 학습 데이터 크기: {2}, 검증 데이터 크기: {3}'
        .format(n_iter, accuracy, train_size, test_size))
  print('#{0} 검증 세트 인덱스:{1}'.format(n_iter, test_index))
  cv_accuracy.append(accuracy)

# 개별 iteration별 정확도를 합하여 평균 정확도 계산
print('\n## 평균 검증 정확도:', np.mean(cv_accuracy))
````
