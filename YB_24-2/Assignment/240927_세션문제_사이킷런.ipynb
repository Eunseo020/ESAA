```python
# 라이브러리 불러오기

import pandas as pd
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
```

```python
train = pd.read_csv('data/train.csv')

# Scailing
scaler = MinMaxScaler()
scaler.fit(train[['fixed acidity']])
train['Scaled fixed acidity'] = scaler.transform(train[['fixed acidity']])

# Encoding
encoder = OneHotEncoder()
encoder.fit(train[['type']])
onehot = encoder.transform(train[['type']])
onehot = onehot.toarray()
onehot = pd.DataFrame(onehot)
onehot.columns = encoder.get_feature_names_out(['type'])
train = pd.concat([train, onehot], axis = 1)
train = train.drop(columns = ['type'])
train.head()
```

4. sklearn에 model_selection 모듈에서 KFold를 불러와보세요

```python
from sklearn.model_selection import KFold
```

5. KFold에 n_splits = 5, shuffle = True, random_state = 0이라는 인자를 추가해 "kf"라는 변수에 저장해보세요

```python
kf = KFold(n_splits=5, shuffle=True, random_state=0)
kf
```

6. 반복문을 통해서 1번부터 5번까지의 데이터에 접근해보세요

```python
for train_idx, val_idx in kf.split(train):
    train_data = train.iloc[train_idx]
    val_data = train.iloc[val_idx]
    #print("train_data:", train_data)
    print('******************')
    print("val index:",val_idx)
    print("val_data:", val_data)
```

```python
# 솔루션-문제와 상관없는 추가코드

import matplotlib.pyplot as plt

kf = KFold(n_splits = 5, shuffle = False) #k=5인 KFold 생성, 섞지 않고 생성
train_idx_store = []  # 빈 train_idx 저장 공간 생성
valid_idx_store = []  # 빈 valid_idx 저장 공간 생성

i = 1
for train_idx, valid_idx in kf.split(train) :   # kf.split으로 train, valid 데이터 나누기
  plt.scatter(valid_idx, [i for x in range(len(valid_idx))], alpha = 0.1) # valid 인덱스마다 y값을 1,2,3,4,5로 설정
  i += 1  # i값 +1
  train_idx_store.append(train_idx)  # 각 폴드의 train_idx 저장
  valid_idx_store.append(valid_idx)  # 각 폴드의 valid_idx 저장
plt.show()
print(train_idx_store)
print(valid_idx_store)

## 아래의 이미지는 shuffle = False로 했을 때, 데이터가 겹치지 않게 순서대로 쪼개지는 것을 볼 수 있음
```
