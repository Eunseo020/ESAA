###**04. 판다스**
- 핵심 객체: DataFrame-여러 개의 행과 열로 이뤄진 2차원 데이터를 담는 구조체. 칼럼dl 여러개인 데이터 구조체(하나는 Series)

(1) 파일을 DataFrame으로 로딩
- read_csv('파일명', sep=','): csv파일 포맷변환해 DataFrame객체로 반환, sep=''에 필드구분 문자 입력, 디폴트는','. / read_table(): 필드구분문자=','/  read_fwf(): 필드구분문자='\t', 고정 길이 기반의 칼럼 포맷
- 맨 왼쪽에 인덱스 생성
- df.head() 디폴트는 맨 앞 5개의 row
- df.shape:행과 열 / df.info(): 총 데이터 건수, 데이터 타입, Null 건수 / df.describe(): 칼럼별 숫자형 데이터의 count(not null건수), mean, std, min, percentile(25%,50%,75%),max, 대략 분포 파악 유용
- df['칼럼']: 칼럼의 Series
- df[].value_counts(): 해당 칼럼의 유형, 데이터값 건수. 인엑스는 고유 칼럼값, 문자열도 가능. dropna=True 디폴트(Null값 제외)

(2) 리스트, 딕셔너리, 넘파이 상호 변환
- 기본적으로 넘파이의 ndarray로 입력이 대부분, 상호 간의 변환 필요 빈번
> ndarray, list, dictionary > DataFrame
 - DataFrame은 칼럼명이 있으므로 칼럼명을 지정함
 - pd.DataFrame(list/array, columns=col_name)
 - pd.DataFrame(ndarray, columns=['','',''])
 - 딕셔너리는 key가 칼럼명으로 매핑됨

 > DataFrame > ndarray, list, dictionary
 -  머신러닝 패키지 입력 인자 적용에 필요
 - df.values: DF를 ndarray로 변환
 - df.values.tolist(): list로 변환
 - df.to_dict('list'): dict로 변환, 인자에 list넣으면 딕셔너리의 값이 리스트형으로 반환


(3) 칼럼 데이터 생성, 수정
- 생성: df['새 칼럼명']=0 or df['기존 칼럼']*10 -상수값 할당 혹은 기본 칼럼 활용
- 수정: df['기존칼럼']=df['기존칼럼']+100

(4) 데이터 삭제
- df.drop(labels=None, axis=0, index=None, cloumns=None, level=None, inplace=False, errors='raise')
 - axis=0: 특정 로우 드롭, axis=1: 특정 칼럼 드롭
 - inplace=False: 원본 df삭제하지 않고 결과df만 반환, inplace=Ture: 원본에서 삭제, 반환값은 None!!
 - df.drop(['칼럼1','칼럼2'], axis=1, inplace=True)


(5) Index
- index 객체는 1차원 ndarray
- df.index: 객체 추출
- df.index.values: 실제값 array로 반환
- indexes[:5].values / indexes.values[:5]
- index 객체는 수정 불가능. 연산함수 적용 제외
- df.rest_index(inplace=False): 새 인덱스를 연속 숫자형(int)으로 할당, 기존인덱스는 'index'라는 새 칼럼 추가, drop=True 시 기존 인덱스 삭제


(6) 데이터 선택, 필터링
- 넘파이: [ ] == 판다스: iloc[ ], loc[ ]
- df['칼럼명'], df[0:2], df[df['P']==3]
- df.iloc[0:2,0]: 위치기반, 행/열 위치
 - df.iloc[:,-1], df.iloc[:,:-1],df.iloc[0:2,[0,1]]
- df.loc['인덱스명','칼럼명']: 명칭기반
 - 슬라이싱 시 종료값 포함
 - 불린 인덱싱 가능 df.loc[df.Year >= 2014]

- 불린 인덱싱: df[], df.loc[]
 - &, |, ~(not)
 - df[(df['A']>6)&(df['P']==1)][['N','B']]


(7) 정렬, Aggregation함수, GroupBy
- df.sort_values(by=['Name','P'],ascending=True,inplace=False) #inplace=True 원본에 적용
- df.min(), df.max(), df.sum(): 모든 칼럼 적용/df.count(): Null 반영안함
- df.groupby(by='Pclass'): 새 객체 반환
 - df.groupby('Pclass').count():Pclass 제외 함수적용  
 - 여러 agg함수 적용: agg()//df.groupby('P')['A'].agg([max,min])
 - agg_format={'A':'max','B':'mean'} 딕셔너리 인자 삽입 가능


(8) 결손 데이터 처리
- df.isna(): NaN 여부 확인(T/F)
 - df.isna().sum(): NaN값 개수
- df.fillna('S',inplace=True): 다른값 대체


(9) apply lambda - 데이터 가공
- lambda x: x**2
 - 인자 여러개일때: a=[1,2,3]; map(lambda x: x**2, a)
 - df['Name_len'].apply(lambda x: len(x)) : 새 칼럼 생성
 - lambda: if/else
   - 반환값, if절: lambda x: 반환값 if x<=15 else 반환값
   - else if: lambda x: '' if x<15 else('' if x<15 else ''))

```python
titanic_df['Child_Adult']=titanic_df['Age'].apply(lambda x: 'Child' if x<=15 else 'Adult')
titanic_df[['Age','Child_Adult']].head(8)
```

```python
titanic_df['Age_cat']=titanic_df['Age'].apply(lambda x: 'Child' if x<=15 else ('Adult' if x<=60 else 'Elderly'))
titanic_df['Age_cat'].value_counts()
```

```python
# 나이에 따라 세분화된 분류를 수행하는 함수 생성,
def get_category(age):
  cat=''
  if age<=5: cat='Baby'
  elif age<=12: cat='Child'
  elif age<=18: cat='Teenager'
  elif age<=25: cat='Student'
  elif age<=35: cat='Young Adult'
  elif age<=60: cat='Adult'
  else: cat='Elderly'

  return cat

titanic_df['Age_cat']=titanic_df['Age'].apply(lambda x: get_category(x))
titanic_df[['Age','Age_cat']].head()
```
